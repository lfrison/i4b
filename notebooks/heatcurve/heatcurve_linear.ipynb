```python
import numpy as np
import pandas as pd
```

```python
def heatcurve_linear(T_amb, T_amb_offset=20, T_sup_offset=20, slope=1):
    T_sup = T_sup_offset - (T_amb - T_amb_offset) * slope
    return T_sup
```

```python
T_amb = np.linspace(-20,20,41)
#T_amb
```

```python
slopes = np.linspace(0,2,11)
T_sup = pd.DataFrame(index=T_amb)
for slope in slopes:
    T_sup[f"slope={slope:.2}"] = heatcurve_linear(T_amb=T_amb, T_amb_offset=20, T_sup_offset=20, slope=slope)
#T_sup.head()
```

```python
T_sup.plot(xlabel="T_amb [degC]", ylabel = "T_sup [degC]")
```
